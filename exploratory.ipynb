```python
import pandas as pd
df = pd.read_spss('/home/bohdan/Downloads/polaczona-baza-danych-z-badania-pracodawcow-bkl-2018-2022.sav')

# %%

df['edycja'] = df['edycja'].astype(int)


def draw_plots(df, year):

  if year is not None:
    df=df[df['edycja']==year]

  #K
  K=df.columns[158:217]
  #R
  R=df.columns[232:352]
  #Z1.1
  Z1=df.columns[373:393]
  #Z2
  Z2=df.columns[353:355]
  #I
  I=df.columns[418:430]

  # %%
  import matplotlib.pyplot as plt

  cols_of_interest=[K,R,Z1,Z2,I]
  cols_of_interest = cols_of_interest
  for col in cols_of_interest:
    nas={}
    for column in col:
      nas[column]=df[column].isna().sum()
    print(nas)
    plt.ylim([0,len(df)])
    plt.title(year)
    plt.bar(nas.keys(), nas.values(), color='g')
    plt.show()

# %%


draw_plots(df, 2018)
draw_plots(df, 2020)
draw_plots(df, 2022)
```

```python
import numpy as np
tmp_df = df.copy()
def calculate_trend(col):
    next_col = f'KP{col[1:]}' 
    tmp_df[col].replace(['w bardzo wysokim stopniu', 'w wysokim stopniu', 'w średnim stopniu', 'w stopniu podstawowym', 'niepotrzebne'], [4, 3, 2, 1, 0], inplace=True)
    tmp_df[next_col].replace(['nie zmieni się', 'trochę straci', 'trochę zyska', 'znacznie straci', 'znacznie zyska'], [0, -1, 1, -2, 2], inplace=True)

    tmp_df[col] = tmp_df[col].dropna().astype(np.int8)
    past = tmp_df[['edycja', col]].dropna().groupby('edycja').median().to_dict()

    tmp_df[next_col] = tmp_df[next_col].dropna().astype(np.int8)
    present = tmp_df[['edycja', next_col]].dropna().groupby('edycja').median().to_dict()
    result = {}

    result['2018'] = past[col][2018] 
    result['2020'] = past[col][2020]
    result['2022'] = past[col][2020] + present[next_col][2022]
    return result


result = []
for col in columns_labels:
    years = calculate_trend(col)
    competence = columns_labels[col]
    result.append({**years, 'competence': competence})
    # print(columns_labels[col], calculate_trend(col))

pd.DataFrame.from_records(result)

# TODO: plot trend
```
